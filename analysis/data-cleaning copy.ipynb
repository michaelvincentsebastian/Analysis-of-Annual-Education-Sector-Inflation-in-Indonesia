```python
import numpy as np

top_header_row_index = 0
bottom_header_row_index = 1
data_start_row_index = 2

top_header = df_combined.iloc[top_header_row_index, :]
bottom_header = df_combined.iloc[bottom_header_row_index, :]
num_cols = df_combined.shape[1]

# --- Membangun Nama Kolom Baru yang Digabungkan ---
new_column_names = []

# Lakukan forward fill pada top_header untuk mengisi NaN
# Jika ada string kosong ('') di header, ganti dulu jadi NaN sebelum ffill
top_header_cleaned = top_header.replace('', np.nan)
top_header_filled = top_header_cleaned.fillna(method='ffill')

for i in range(num_cols):
    if i == 0:
        new_column_names.append('Wilayah')
    elif i == num_cols - 1: # Kolom terakhir, kemungkinan Tahun
        # Cek jika nilainya berupa angka di salah satu header (untuk kolom Tahun)
        if (pd.isna(top_header[i]) and str(bottom_header[i]).strip().isdigit()) or \
           (str(top_header[i]).strip().isdigit() and pd.isna(bottom_header[i])):
            new_column_names.append('Tahun')
        else: # Kasus lain untuk kolom terakhir, mungkin Tahunan summary
            category = str(top_header_filled[i]).strip() if not pd.isna(top_header_filled[i]) else ""
            month_or_summary = str(bottom_header[i]).strip() if not pd.isna(bottom_header[i]) else ""
            if category and month_or_summary:
                new_column_names.append(f"{category}_{month_or_summary}")
            elif category:
                new_column_names.append(f"{category}_Summary")
            else:
                new_column_names.append(f"Unnamed_Col_{i}")
    else: # Kolom data inflasi bulanan
        category = str(top_header_filled[i]).strip() if not pd.isna(top_header_filled[i]) else ""
        month_or_summary = str(bottom_header[i]).strip() if not pd.isna(bottom_header[i]) else ""

        if category and month_or_summary:
            new_column_names.append(f"{category}_{month_or_summary}")
        elif category and not month_or_summary:
             new_column_names.append(f"{category}_Summary")
        elif not category and month_or_summary: # Jika kolom bulan tidak ada kategori atas (seperti kolom pertama 'Januari' setelah 'Wilayah')
             new_column_names.append(f"Unnamed_Kategori_{month_or_summary}")
        else: # Kolom yang benar-benar kosong di kedua level header
            new_column_names.append(f"Unnamed_Col_{i}")

# --- Terapkan Nama Kolom dan Ambil Data ---
df_cleaned = df_combined.iloc[data_start_row_index:].copy()
df_cleaned.columns = new_column_names

# --- Hapus Kolom yang Tidak Relevan ---
cols_to_drop = [col for col in df_cleaned.columns if 'Unnamed_Col_' in col or 'Unnamed_Kategori_' in col]
df_cleaned = df_cleaned.drop(columns=cols_to_drop, errors='ignore')

# --- Tangani Missing Values ('-') dan Konversi Tipe Data ---
df_cleaned = df_cleaned.replace('-', np.nan)
inflation_cols = [col for col in df_cleaned.columns if col not in ['Wilayah', 'Tahun']]

for col in inflation_cols:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

# --- Unpivoting (Mengubah dari Wide ke Long/Tidy Format) ---
id_vars_cols = ['Wilayah', 'Tahun']
value_vars_cols = [col for col in df_cleaned.columns if col not in id_vars_cols]

df_tidy = pd.melt(df_cleaned,
                  id_vars=id_vars_cols,
                  value_vars=value_vars_cols,
                  var_name='Kategori_Bulan',
                  value_name='Nilai_Inflasi')

# --- Pisahkan Kolom 'Kategori_Bulan' menjadi 'Kategori' dan 'Bulan' ---
def split_category_month_v2(col_name):
    if col_name.endswith('_Summary'):
        return col_name.replace('_Summary', ''), 'Tahunan'
    parts = col_name.rsplit('_', 1)
    if len(parts) == 2:
        return parts[0], parts[1]
    return col_name, np.nan

df_tidy[['Kategori_Pendidikan', 'Bulan']] = df_tidy['Kategori_Bulan'].apply(split_category_month_v2).apply(pd.Series)
df_tidy = df_tidy.drop(columns=['Kategori_Bulan'])

# Urutkan untuk tampilan yang lebih rapi
bulan_order = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember', 'Tahunan']
df_tidy['Bulan'] = pd.Categorical(df_tidy['Bulan'], categories=bulan_order, ordered=True)

kategori_order = [
    'Pendidikan',
    'Pendidikan Dasar dan Anak Usia Dini',
    'Pendidikan Menengah',
    'Pendidikan Tinggi',
    'Pendidikan Lainnya'
]
df_tidy['Kategori_Pendidikan'] = df_tidy['Kategori_Pendidikan'].astype(str)
df_tidy['Kategori_Pendidikan'] = pd.Categorical(df_tidy['Kategori_Pendidikan'], categories=kategori_order, ordered=True)

df_tidy = df_tidy.sort_values(by=['Wilayah', 'Tahun', 'Kategori_Pendidikan', 'Bulan']).reset_index(drop=True)

# Tampilkan hasil akhir
print(df_tidy.head(20))
```
