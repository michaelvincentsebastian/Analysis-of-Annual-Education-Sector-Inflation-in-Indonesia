```python
df24 = pd.read_excel(r'B:\GitHub Repository\Analysis-of-Annual-Education-Sector-Inflation-in-Indonesia\datasets\2024.xlsx', header=None, skiprows=2)
df25 = pd.read_excel(r'B:\GitHub Repository\Analysis-of-Annual-Education-Sector-Inflation-in-Indonesia\datasets\2025.xlsx', header=None, skiprows=2)
```

Make new column 'Tahun' and fill it by the dataframe year

```python
df24['Tahun'] = 2024
df25['Tahun'] = 2025
```

Drop the Year Column in the dataframe

```python
df25.drop(df25.index[1], inplace=True)
df24.drop(df24.index[1], inplace=True)
```

Fill the NaN in 2024 dataframe with ''

```python
df24.fillna('', inplace=True)
```

Remove the Header of 2025 dataframe, so that only contains the values. This will make the concat clean.

```python
df25.reset_index(drop=True, inplace=True)
df25 = df25.iloc[2:]
df25.reset_index(drop=True, inplace=True)
df25
```

Join the 2024 df and 2025 df with concat method

```python
df_concat = pd.concat([df24, df25], ignore_index=True)
df_concat.reset_index(drop=True, inplace=True)
df_concat
```

Create Multi index in concat dataframe from index 0 to 1

```python
# Create MultiIndex columns from first two rows
df_concat.columns = pd.MultiIndex.from_arrays([df_concat.iloc[0], df_concat.iloc[1]])

# Drop the first two rows since they're now part of the column headers
df_concat = df_concat.iloc[2:].reset_index(drop=True)

# Convert numeric columns to float where possible
for col in df_concat.columns:
    try:
        df_concat[col] = df_concat[col].replace('', np.nan).astype(float)
    except:
        continue
```
