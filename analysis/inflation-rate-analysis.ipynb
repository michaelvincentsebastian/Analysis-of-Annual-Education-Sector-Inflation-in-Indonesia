2. Data Cleaning

- Menghapus Baris yang mengandung Duplicate

```python
df24.drop_duplicates(inplace=True)
df25.drop_duplicates(inplace=True)
```

- Menetapkan index 3 sebagai awal dari Data Frame

```python
df24 = df24.iloc[3:]
df25 = df25.iloc[3:]
df24
df25
```

- Mereset Index (Karena sebelumnya index awal kita masih 3 belum 0)

```python
df24.reset_index(drop=True, inplace=True)
df25.reset_index(drop=True, inplace=True)
```

```python
df24
```

```python
df25
```

- Mengisi Semua NaN menjadi '' atau kosong

```python
# Replace NaN with empty string
df24.fillna('', inplace=True)
df25.fillna('', inplace=True)
```

- Mengatur index 0 (Bulan Januari, Februari, dst.) menjadi Columns Header nya

```python
# Set row with index 0 (containing Januari, Februari, etc.) as header
df24.columns = df24.iloc[0]
df25.columns = df25.iloc[0]
```

- Karena index 0 sudah kita set sebagai Header maka kita perlu menghapus index 0

```python
# Drop the row that became header
df24 = df24.iloc[1:]
df25 = df25.iloc[1:]
```
