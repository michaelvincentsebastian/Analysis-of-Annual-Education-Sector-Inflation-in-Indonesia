```python
# Convert numeric columns to float
for col in pendidikan.columns[1:]:
    pendidikan[col] = pd.to_numeric(pendidikan[col], errors='coerce')

# Bersihkan baris tidak perlu
pendidikan = pendidikan.iloc[1:]
pendidikan = pendidikan.dropna(how='all', subset=pendidikan.columns[1:])

# Nama bulan
month_names = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 
               'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

# Ambil data maksimum tiap bulan
result = []

for month_num, month_name in enumerate(month_names, 1):
    max_inflation = pendidikan[month_num].max()
    region = pendidikan.loc[pendidikan[month_num] == max_inflation, 'Region'].iloc[0]
    result.append([month_name, region, round(max_inflation, 2)])

# DataFrame hasil
max_inflasi_df = pd.DataFrame(result, columns=['Bulan', 'Region', 'Inflasi'])

# Warna
warna_header = "#4A9782"
warna_cell = "#DDF6D2"

# Visualisasi
fig, ax = plt.subplots(figsize=(10, 5))
ax.axis('off')

# Buat tabel
tabel = ax.table(cellText=max_inflasi_df.values,
                 colLabels=max_inflasi_df.columns,
                 cellLoc='center',
                 loc='center')

# Styling
tabel.auto_set_font_size(False)
tabel.set_fontsize(10)
tabel.scale(1, 1.5)

# Pewarnaan cell
for key, cell in tabel.get_celld().items():
    row, col = key
    if row == 0:  # header
        cell.set_facecolor(warna_header)
        cell.set_text_props(weight='bold', color='white')
    else:
        cell.set_facecolor(warna_cell)  # semua data

plt.tight_layout()
plt.show()
```

```python
# Display data for KOTA BIMA in Juli and Agustus
kota_bima_data = pendidikan[pendidikan['Region'] == 'KOTA BIMA'][[7, 8]]
print("KOTA BIMA data for Juli and Agustus:")
print(kota_bima_data)

# Also show the highest inflation values for these months to verify
juli_max = pendidikan[pendidikan[7] == pendidikan[7].max()]
agustus_max = pendidikan[pendidikan[8] == pendidikan[8].max()]

print("\nHighest inflation in Juli:")
print(juli_max[['Region', 7]])

print("\nHighest inflation in Agustus:")
print(agustus_max[['Region', 8]])
```

```python
# Convert numeric columns to float
for col in pendidikan.columns[1:]:
    pendidikan[col] = pd.to_numeric(pendidikan[col], errors='coerce')

# Bersihkan baris tidak perlu
pendidikan = pendidikan.iloc[1:]
pendidikan = pendidikan.dropna(how='all', subset=pendidikan.columns[1:])

# Nama bulan
month_names = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 
               'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

# Ambil data minimum tiap bulan (khusus deflasi = negatif)
result = []

for month_num, month_name in enumerate(month_names, 1):
    min_inflation = pendidikan[month_num].min()
    if min_inflation < 0:
        region = pendidikan.loc[pendidikan[month_num] == min_inflation, 'Region'].iloc[0]
        result.append([month_name, region, round(min_inflation, 2)])

# DataFrame hasil deflasi
deflasi_df = pd.DataFrame(result, columns=['Bulan', 'Region', 'Deflasi'])

# Warna
warna_header = "#DC3C22"
warna_cell = "#FFD8D8"

# Visualisasi
fig, ax = plt.subplots(figsize=(10, 5))
ax.axis('off')

# Buat tabel
tabel = ax.table(cellText=deflasi_df.values,
                 colLabels=deflasi_df.columns,
                 cellLoc='center',
                 loc='center')

# Styling
tabel.auto_set_font_size(False)
tabel.set_fontsize(10)
tabel.scale(1, 1.5)

# Pewarnaan cell
for key, cell in tabel.get_celld().items():
    row, col = key
    if row == 0:  # header
        cell.set_facecolor(warna_header)
        cell.set_text_props(weight='bold', color='white')
    else:
        cell.set_facecolor(warna_cell)  # semua data

plt.tight_layout()
plt.show()
```

```python
# Convert numeric columns for each sector
for df_sector in [pendidikan, pendidikan_dasar, pendidikan_menengah, pendidikan_tinggi, pendidikan_lainnya]:
    for col in df_sector.columns[1:]:
        df_sector[col] = pd.to_numeric(df_sector[col], errors='coerce')

# Calculate mean inflation for each sector
sector_means = {
    'Pendidikan': pendidikan.iloc[:, 1:].mean().mean(),
    'Pendidikan Dasar': pendidikan_dasar.iloc[:, 1:].mean().mean(),
    'Pendidikan Menengah': pendidikan_menengah.iloc[:, 1:].mean().mean(),
    'Pendidikan Tinggi': pendidikan_tinggi.iloc[:, 1:].mean().mean(),
    'Pendidikan Lainnya': pendidikan_lainnya.iloc[:, 1:].mean().mean()
}

# Create bar plot with green color scheme
plt.figure(figsize=(12, 6))
colors = ['#2D5A27', '#4A9782', '#8FBC94', '#C5E5A4', '#DDF6D2']
bars = plt.bar(sector_means.keys(), sector_means.values(), color=colors)

# Customize plot
plt.title('Average Inflation Comparison Across Education Sectors', pad=20, fontsize=12)
plt.ylabel('Average Inflation Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%',
             ha='center', va='bottom')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate mean deflation (negative inflation) for each sector
sector_deflation = {
    'Pendidikan': pendidikan.iloc[:, 1:][pendidikan.iloc[:, 1:] < 0].mean().mean(),
    'Pendidikan Dasar': pendidikan_dasar.iloc[:, 1:][pendidikan_dasar.iloc[:, 1:] < 0].mean().mean(),
    'Pendidikan Menengah': pendidikan_menengah.iloc[:, 1:][pendidikan_menengah.iloc[:, 1:] < 0].mean().mean(),
    'Pendidikan Tinggi': pendidikan_tinggi.iloc[:, 1:][pendidikan_tinggi.iloc[:, 1:] < 0].mean().mean(),
    'Pendidikan Lainnya': pendidikan_lainnya.iloc[:, 1:][pendidikan_lainnya.iloc[:, 1:] < 0].mean().mean()
}

# Create bar plot with red color for deflation
plt.figure(figsize=(12, 6))
bars = plt.bar(sector_deflation.keys(), sector_deflation.values(), color='#DC3C22')
plt.title('Average Deflation Comparison Across Education Sectors', pad=20)
plt.ylabel('Average Deflation Rate (%)')
plt.xticks(rotation=45)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%',
             ha='center', va='bottom')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate monthly averages for Pendidikan
monthly_avg = pendidikan.iloc[:, 1:].mean()

# Create month labels
month_names = ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 
               'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember']

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(month_names, monthly_avg.values, color='#06923E', marker='o', linewidth=2, markersize=8)

# Customize the plot
plt.title('Average Monthly Education Inflation', pad=20, fontsize=12)
plt.ylabel('Average Inflation Rate (%)')
plt.xlabel('Month')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

# Add value labels on top of each point
for i, value in enumerate(monthly_avg.values):
    plt.text(i, value, f'{value:.2f}%', ha='center', va='bottom')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Calculate average inflation for each region across all months
pendidikan['avg_inflation'] = pendidikan.iloc[:, 1:].mean(axis=1)

# Get Indonesia's average
indonesia_avg = pendidikan[pendidikan['Region'] == 'INDONESIA']['avg_inflation'].values[0]

# Get top 3 regions (excluding Indonesia)
top_3_regions = pendidikan[pendidikan['Region'] != 'INDONESIA'].nlargest(3, 'avg_inflation')

# Prepare data for plotting
regions = ['Indonesia'] + list(top_3_regions['Region'])
values = [indonesia_avg] + list(top_3_regions['avg_inflation'])

# Create bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(regions, values, color='#06923E')

# Customize plot
plt.title('Average Education Inflation: Indonesia vs Top 3 Regions', pad=20)
plt.ylabel('Average Inflation Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%',
             ha='center', va='bottom')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
